```python
def list_maker2():
    lis=[]
    for i in range(1,11):
        lis.append("Unnamed: "+str(i*11))
    return lis
```

```python
def clean(Dat_f):                            #DO NOT CALL MULTIPLE TIMES
    lis=range(0,8)
    Dat_f.drop(lis, axis=0, inplace=True)    #clearing first 6 rows 
    lis=list_maker()
    Dat_f.drop(lis, axis=1, inplace=True)    #clearing error columns
    lis=list_maker2()
    lis.append("Table with row headers in column A, L, W, AH, AS, BD, BO, BZ, CK, CV, and DG, and column headers in rows 6 through 8 and 46 through 48.")
    Dat_f.drop(lis, axis=1, inplace=True)    #clearing visibility label columns
    lis=[13,19,25,31,37,43,44,45,46,47,53,59,65,71,74,75,76]
    Dat_f.drop(lis, axis=0, inplace=True)    #clearing misc rows
    try:
        Dat_f.drop([77],axis=0,inplace=True)
    except:
        print("🤞")
    #return Dat_f
```

```python
def clean2(df):                              #CAN CALL Multiple times
    arr=df.to_numpy()
    np.fill_diagonal(arr, 0)
    for i in range(0,52):
        for j in range(0,52):
            arr[i,j]=int(str(arr[i,j]).replace(',', ''))
    #label=range(0,52)
    label=["Alabama","Alaska","Arizona","Arkansas","California","Colorado","Connecticut","Delaware","District of Columbia","Florida","Georgia","Hawaii","Idaho","Illinois","Indiana","Iowa","Kansas","Kentucky","Louisiana","Maine","Maryland","Massachusetts","Michigan","Minnesota","Mississippi","Missouri","Montana","Nebraska","Nevada","New Hampshire","New Jersey","New Mexico","New York","North Carolina","North Dakota","Ohio","Oklahoma","Oregon","Pennsylvania","Puerto Rico","Rhode Island","South Carolina","South Dakota","Tennessee","Texas","Utah","Vermont","Virginia","Washington","West Virginia","Wisconsin","Wyoming"]
    return pd.DataFrame(arr,index=label, columns=label)
    
def dat_conv(df):
    for column in df.columns:
        df[column] = df[column].astype(str).astype(int)
```

```python
def main(df,n):
    for i in range(5,n+1):
        print(3)
        j=str(i)
        j= j + '.csv'
        print(j)
        df.append(pd.DataFrame(pd.read_csv(j))) 
        clean(df[i-5])
        df[i-5]=clean2(df[i-5])
        dat_conv(df[i-5])
        print(2)
        df[i-5].to_csv('Cleansheet'+j+'.csv')
        print(1)

    return df
    
    
df=[]
df=main(df,19)
```
