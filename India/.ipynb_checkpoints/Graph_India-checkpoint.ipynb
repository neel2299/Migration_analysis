```python
import pandas as pd
import numpy as np
import networkx as nx
```

```python
df=pd.read_excel("2011_R_U.XLSX")
for index, row in df.iterrows():
    df.iloc[index,1]=row["To"][8:-5]
    x=row["From"]
    df.iloc[index,0]=x.upper()
for index,row in df.iterrows():
    if row["From"]=="TOTAL":
        df.drop(index,axis=0,inplace=True)
```

```python
df

lis=[]
for index, row in df.iterrows():
    lis.append((row["From"],row["To"],row["Persons"]))
    
lis
```

```python
df.to_csv("2011T.csv",header=None,index=False)
df = pd.read_csv('2011T.csv', delim_whitespace = True, 
                   header = None, names =['n1', 'n2', 'weight'])
G = nx.DiGraph()
G.add_weighted_edges_from(lis)
```
