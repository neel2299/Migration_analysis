```python
r = requests.get('https://m.rbi.org.in/Scripts/PublicationsView.aspx?id=20002')
soup = BeautifulSoup(r.content, 'html.parser')
data = []
table = soup.find('table', attrs={'class':'lineItemsTable'})


rows = soup.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele])

dic_urbanemp={}
itr=85
lenn=len(data)
#for i in range(5)
while itr<120:
    try:
        dic_urbanemp[data[itr][0].upper()]=float(data[itr][5])/10
    except ValueError:
        dic_urbanemp[data[itr][0].upper()]=1.7
   # print(dic_urbanemp[data[itr][0].upper()])
    itr+=1
dic_urbanemp["DELHI NCT"]=dic_urbanemp["DELHI"]
dic_urbanemp["ANDAMAN & NICOBAR ISLANDS"]=dic_urbanemp["ANDAMAN AND NICOBAR ISLANDS"]
```

```python
final_dic={}                           #Net unemployment dict
for key in dic_urbanemp.keys():
    var_mid=(dic_ruralemp[key]*dic_ruralpop[key]+dic_urbanemp[key]*dic_urbanpop[key])/(dic_ruralpop[key]+dic_urbanpop[key])
    final_dic[key] = float("{0:.2f}".format(var_mid))
final_dic['JAMMU AND KASHMIR']=final_dic["JAMMU & KASHMIR"]
```

```python
import networkx as nx
import matplotlib.pyplot as plt
for issac in asimov:
    issac=issac[:-3]+"gml"
    G=nx.read_gml(issac)
    G.remove_edges_from(nx.selfloop_edges(G))

    dic1=dic_invest
    dic2=dic_invest
    y=0
    count=0
    x=[]
    for u,v,a in G.edges(data=True):
        x+=[round((dic2[v]-dic1[u]),1)]*(a["weight"])
        count=count+a["weight"]
        y=round((dic2[v]-dic1[u]),1)*a["weight"]+y
    plt.hist(x, histtype='bar', )
    strr="average trend: "+str(round((y/count),1))
    #plt.text(25000, 35.0*1e4, strr)
    plt.title("("+issac[:-4]+")"+strr)
    plt.ylabel("Net migrants")
    plt.xlabel("Destination investment-Origin investment")
    fig=plt.gcf()
    fig.savefig(issac[:-4]+"_INV")
    plt.show()
    len(x)
```
